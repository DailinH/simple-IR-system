# construct inverted index
documents to be indexed => tokenize => modify tokens => inverted index

## document sources
1.txt : all is well that ends well  
2.txt : as you like it  
3.txt : the comedy of errors  
4.txt : cymbeline  

```python
# read file
doc1 = open('./text/1.txt','r')
doc2 = open('./text/2.txt','r')
doc3 = open('./text/3.txt','r')
doc4 = open('./text/4.txt','r')

# indexer steps: sequence of (modified token, dcument ID)pairs
seq = []
def generate_seq(doc,num):
    for i, line in enumerate(doc):
        line = line[:-1].replace('.','').replace(':','').replace(',','') #delete the final endline \n
        for tk in line.split(' '): #split with ' '
            if tk != '':
                seq.append([tk.lower(),num])
            
generate_seq(doc1,1)
generate_seq(doc2,2)
generate_seq(doc3,3)
generate_seq(doc4,4)

print(len(seq))
print(seq)
```

```python
# sort by terms
sorted_tk = sorted(seq,key= lambda x : x[0])
print(len(sorted_tk))
print(sorted_tk)
```

```python
# add frequency information
# (term, doc #, term freq)
freq_info = []
for i in sorted_tk:
    if len(freq_info) == 0:
        freq_info.append([i[0],i[1],1])
        continue
    if i == freq_info[-1][:2]:
        freq_info[-1][2]+=1
    else:
        freq_info.append([i[0],i[1],1])
print(len(freq_info))
print(freq_info)
```

```python
# split into dictionary file & posting file
# dictionary: (term, N docs, col freq)
# posting: (doc, freq)
dictionary_file = []
posting_file = [x[1:] for x in freq_info]

for i in freq_info:
    if len(dictionary_file) == 0:
        dictionary_file.append([i[0], 1, i[2]])
        continue
    if dictionary_file[-1][0] == i[0]:
        dictionary_file[-1][1]+=1
        dictionary_file[-1][2] += i[2]
    else:
        dictionary_file.append([i[0],1,i[2]])
print(len(dictionary_file))
print(dictionary_file)
```

```python
# query processing
def get_ptr(tk):
    """returns the starting point in posting list for token tk"""
    ans = 0
    for i in dictionary_file:
        if i[0] != tk:
            ans+=i[1]
        else:
            return posting_file[ans:ans+i[1]]
        
# print(get_ptr('a'))
        
def AND(a, b):
    """
    check when two tokens a and b both appear
    """
    ans = []
    a_list = get_ptr(a)
    b_list = get_ptr(b)

    a_i = 0
    b_i = 0
    while a_i < len(a_list) and b_i < len(b_list):
#         print(a_list[a_i])
        if(a_list[a_i][0] == b_list[b_i][0]):
            ans.append(a_list[a_i][0])
            a_i +=1
            b_i +=1
        elif a_list[a_i][0] < b_list[b_i][0]:
            a_i +=1
        else:
            b_i +=1
    return ans

print(AND('a','abbey'))
```
